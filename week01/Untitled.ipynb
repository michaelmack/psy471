# Our first model
## A simple theory
What is memory? Remembering things from the past, right? But, to remember something, you had to put that something in memory in the first place. So, to have a memory, we need to be able to store experiences into memory (i.e., memory encoding) and then retrieve those experiences from memory (i.e., memory retrieval). That means you go through life experiencing different events which are stored into memory through encoding functions and then later remembered through retrieval functions. That sounds like memory!
