# Variables and standard math operations
Python has the usual suspects in terms of math operations. A tricky one is division, since the output you get from a division call depends on the data type of the variables you are using (remember this from HW Zero?).

```python
a = 5
b = 4
print a+b
print a-b
print a*b
print a/b
print a/(b*1.0)
```

```python
c = 2.4
d = 1.5
f = c*d
print f
```

# Lists
Lists allow you to store multiple values in one variable.

```python
g = [1,2,3]
print g
```

Items can be added to list with the `append` function

```python
g.append(28)
print g
```

```python
print g[0]
```

You can change the value at a specific location in a list by using its index.

# Numpy arrays
Numpy is a python library that provides a host of numerical functions and data types that make computations much easier. One critical aspect of numpy is its arrays. Arrays are similar to lists, but can handle elementwise computations.

```python
# here's a list
a = [1,2,3]
# what do you think is going to happen?
b = a*3
print b
```

Probably not what you had in mind. Using the multiplication operator with a list creates a new list that repeats the original list the specified number of times. Let's try a numpy array instead.

```python
# here's a numpy array
c = np.array([1,2,3])
d = c*3
print d
```

That's better. Note the use of the shorthand `np` before the array function. Way back at the top, I imported numpy and gave it a shorthand name of np (e.g., `import numpy as np`). This lets me use `np` instead of `numpy` to access numpy's functions.

Numpy arrays are indexed in the same way as lists:

```python
print d[2]
d[0] = d[0]+2
print d
```

Broadcasting is a useful way to index the value in multiple locations of a numpy arrays:

```python
a = np.array([0,1,2,3,4,5,6])
print a[2:5] # elements from index 2 to index 5 (but don't include value at index 5)
print a[4:] # elements from index 4 to end of array
print a[:3] # elements from beginning to index 3 (but don't include value at index 3)
```

Numpy provides several functions for creating different kinds of arrays:

```python
a = np.zeros((2,2))  # Create an array of all zeros
print a              # Prints "[[ 0.  0.]
                     #          [ 0.  0.]]"

b = np.ones((1,2))   # Create an array of all ones
print b              # Prints "[[ 1.  1.]]"

c = np.full((2,2), 7) # Create a constant array
print c               # Prints "[[ 7.  7.]
                      #          [ 7.  7.]]"

d = np.eye(2)        # Create a 2x2 identity matrix
print d              # Prints "[[ 1.  0.]
                     #          [ 0.  1.]]"
    
e = np.random.random((2,2)) # Create an array filled with random values
print e  
```

```python
# indexing into a multidimensional array
print e[0][1]
```

```python
# array operations
print c+e
print c/e
print c*e
```
